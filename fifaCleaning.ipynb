```python
fifa_df = fifa_df.drop(['LongName', 'photoUrl', 'playerUrl', 'Loan Date End'], axis=1)
```

```python
# Drop NaN values
fifa_df.dropna()
```

```python
# Drop find total null values
fifa_df.isnull().sum()
```

```python
# Fill 'Hit' column with '0' for null values
fifa_df['Hits'] = fifa_df['Hits'].fillna(0)
```

## Convert the height and weight columns to numerical forms


```python
fifa_df['Height'].unique()
```

```python
# Remove 'cm' from Height column values and convert feet/inches to centimeters
height = []

for i in range(len(fifa_df)):
    Height = fifa_df['Height'][i]
    
    if 'cm' in Height:
        Height = float(Height.rstrip('cm'))
        height.append(Height)
        
    elif "'" in Height: 
        Height = Height.rstrip('"').split("'")
        Height = float(Height[0]) * 12 + float(Height[1])
        Height = Height * 2.54
        Height = round(Height, 2)        
        height.append(Height)

fifa_df['Height'] = height
```

```python
fifa_df['Height'].unique()
```

```python
fifa_df['Weight'].unique()
```

```python
# Remove 'kg' or 'lbs' from Weight column
weight = []
for i in range(len(fifa_df)):
    Weight = fifa_df['Weight'][i]
    
    # Convert values from 'kg' to 'lbs'
    if 'kg' in Weight:
        Weight = float(Weight.rstrip('kg')) * 2.20462
        Weight = round(Weight, 2)
        weight.append(Weight)
    else:
        Weight = float(Weight.rstrip('lbs'))
        weight.append(Weight)

fifa_df['Weight'] = weight
```

```python
fifa_df['Weight'].unique()
```

### Remove the unnecessary newline characters from all columns that have them.


```python
fifa_df.head()
```

```python
# Remove '\n' from 'Club' column
fifa_df['Club'] = fifa_df['Club'].replace(r'\s+|\\n', ' ', regex=True) 
fifa_df.head()
```

### Based on the 'Joined' column, check which players have been playing at a club for more than 10 years!




```python
fifa_df['Joined'] = fifa_df['Joined'].str.replace(",", "").str.split(" ")
```

```python
fifa_df['Joined']
```

```python
# Convert month name to month number
correct_date = []

for i in range(len(fifa_df['Joined'])):
    month = fifa_df['Joined'][i][0]
    day = fifa_df['Joined'][i][1]
    year = fifa_df['Joined'][i][2]    
    
    if month == 'Jan':
        month = '1'
    elif month == 'Feb':
        month = '2'
    elif month == 'Mar':
        month = '3'
    elif month == 'Apr':
        month = '4'
    elif month == 'May':
        month = '5'        
    elif month == 'Jun':
        month = '6'
    elif month == 'Jul':
        month = '7'
    elif month == 'Aug':
        month = '8'
    elif month == 'Sep':
        month = '9'
    elif month == 'Oct':
        month = '10'
    elif month == 'Nov':
        month = '11'
    else:
        month = '12'
        
    if len(day) == 1:
        day = ('0' + str(day))
    
    date_format = (month + '-' + day + '-' + year)
    date_format = pd.to_datetime(date_format)
    correct_date.append(date_format)
    
fifa_df['Joined'] = correct_date
```

```python
time_at_club = []
for i in range(len(fifa_df)):
    time_at_club.append(pd.Timestamp.now().year - correct_date[i].year)

fifa_df['Playing at Club (yrs)'] = time_at_club
```

### 'Value', 'Wage' and "Release Clause' are string columns. Convert them to numbers. For eg, "M" in value column is Million, so multiply the row values by 1,000,000, etc.

```python
fifa_df['Value'].unique()
```

```python
# Remove '€' sign and 'M'/'K' from columns
value = []
for i in range(len(fifa_df)):
    Value = fifa_df['Value'][i]
    
    # Convert strings to their respective multiple
    if 'M' in Value:
        Value = Value.replace("€","").rstrip("M")
        Value = pd.to_numeric(Value, errors='coerce')
        value.append(Value * 1000000)
    else:
        Value = Value.replace("€","").rstrip("K")
        Value = pd.to_numeric(Value, errors='coerce')
        value.append(Value * 1000) 

fifa_df['Value'] = value        
```

```python
# Remove '€' sign and 'M'/'K' from columns
wage = []
for i in range(len(fifa_df)):
    Wage = fifa_df['Wage'][i]
    
    # Convert strings to their respective multiple
    if 'M' in Wage:
        Wage = Wage.replace("€","").rstrip("M")
        Wage = pd.to_numeric(Wage, errors='coerce')
        wage.append(Wage * 1000000)
    else:
        Wage = Wage.replace("€","").rstrip("K")
        Wage = pd.to_numeric(Wage, errors='coerce')
        wage.append(Wage * 1000) 

fifa_df['Wage'] = wage        
```

```python
# Remove '€' sign and 'M'/'K' from columns
release_clause = []
for i in range(len(fifa_df)):
    Release = fifa_df['Release Clause'][i]
    
    # Convert strings to their respective multiple
    if 'M' in Release:
        Release = Release.replace("€","").rstrip("M")
        Release = pd.to_numeric(Release, errors='coerce')
        release_clause.append(Release * 1000000)
    else:
        Release = Release.replace("€","").rstrip("K")
        Release = pd.to_numeric(Release, errors='coerce')
        release_clause.append(Release * 1000) 

fifa_df['Release Clause'] = release_clause        
```

```python
# Convert 'Hits' column from object to integer
hits = []
for i in range(len(fifa_df)):
    Hits = fifa_df['Hits'][i]
    
    if type(Hits) == str and 'K' in Hits:
        Hits = Hits.replace('K', '')
        Hits = float(Hits) * 1000
        hits.append(Hits)
    else:
        hits.append(Hits)

fifa_df['Hits'] = hits
```
